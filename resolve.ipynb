```python
import pandas as pd

data = {
    "complaint": [
        "Payment failed but money deducted",
        "App is crashing again and again",
        "Delivery is very late",
        "Received damaged product",
        "Refund not processed yet",
        "Login issue, unable to access account",
        "Wrong item delivered",
        "Payment issue again, very frustrated"
    ],
    "category": [
        "Payment Issue",
        "Technical Issue",
        "Delivery Issue",
        "Product Issue",
        "Payment Issue",
        "Technical Issue",
        "Product Issue",
        "Payment Issue"
    ],
    "urgency": [
        "High",
        "Medium",
        "Low",
        "Medium",
        "High",
        "High",
        "Medium",
        "High"
    ]
}

df = pd.DataFrame(data)
print(df.head())
```

```python
import nltk
from nltk.corpus import stopwords
import string

nltk.download('stopwords')

stop_words = set(stopwords.words('english'))

def preprocess(text):
    text = text.lower()
    text = "".join([c for c in text if c not in string.punctuation])
    words = text.split()
    words = [w for w in words if w not in stop_words]
    return " ".join(words)

df["clean_text"] = df["complaint"].apply(preprocess)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(df["clean_text"])
```

```python
from sklearn.linear_model import LogisticRegression

cat_model = LogisticRegression()
cat_model.fit(X, df["category"])
```

```python
urg_model = LogisticRegression()
urg_model.fit(X, df["urgency"])
```

```python
def predict_complaint(text):
    clean = preprocess(text)
    vec = vectorizer.transform([clean])

    category = cat_model.predict(vec)[0]
    urgency = urg_model.predict(vec)[0]

    # Rule-based action suggestion
    if urgency == "High":
        action = "Immediate escalation"
    elif urgency == "Medium":
        action = "Handle within 24 hours"
    else:
        action = "Normal processing"

    return {
        "Category": category,
        "Urgency": urgency,
        "Suggested Action": action
    }
```

```python
test = "My payment failed but money was deducted twice"

result = predict_complaint(test)

for k, v in result.items():
    print(f"{k}: {v}")
```
